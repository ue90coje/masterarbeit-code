```python
from python_files.ResultsIOHandler import ResultsIOHandler, compare_results_in_table
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix, classification_report
import matplotlib.pyplot as plt
from python_files.Vorverarbeitung import Szenario
```

```python
results_handler_train = ResultsIOHandler(Szenario.Szenario1, False)
results_handler_train_and_test = ResultsIOHandler(Szenario.Szenario1, True)
```

```python
results_train = results_handler_train.get_results()
results_train_and_test = results_handler_train_and_test.get_results()

#Ausgabe der aller Ergebnisse mit Accuracy und F1-Score
print("Results for training data:")
for result in results_train_and_test:
    true_labels = result["true_labels"]
    probas = result["probas"]
    predictions = probas > 0.5
    print("Spalten: ", result["column_combination"])
    print("ROC AUC: ", roc_auc_score(true_labels, probas))
    print("Confusion Matrix: \n", confusion_matrix(true_labels, predictions))
    print("Classification Report: \n", classification_report(true_labels, predictions))
```
