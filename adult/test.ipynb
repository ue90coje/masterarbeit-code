```python
import pandas as pd
import matplotlib.pyplot as plt

# Definiere die übergeordneten und untergeordneten Kategorien
kategorien_ober = ['Kategorie A', 'Kategorie B', 'Kategorie C']
kategorien_unter = ['Unterkategorie 1', 'Unterkategorie 2', 'Unterkategorie 3']

# Erzeuge die MultiIndex-Spalten
columns = pd.MultiIndex.from_product([kategorien_ober, kategorien_unter])

# Erzeuge Beispieldaten für die Tabelle
data = [
    [1, 2, 3, 4, 5, 6, 7, 8, 9],
    [9, 8, 7, 6, 5, 4, 3, 2, 1],
    [2, 3, 4, 5, 6, 7, 8, 9, 1]
]

# Erstelle den DataFrame mit den hierarchischen Spalten
df = pd.DataFrame(data, columns=columns)

# Plotten der Tabelle
fig, ax = plt.subplots(figsize=(10, 4))
ax.axis('off')  # Achsen deaktivieren
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

# Optische Anpassungen (optional)
table.auto_set_font_size(False)
table.set_fontsize(18)
table.scale(2, 2)

plt.show()
```
