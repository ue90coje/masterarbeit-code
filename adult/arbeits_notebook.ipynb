```python
import pandas as pd
from python_files.Data_Loader import Data_Loader
from python_files.Vorverarbeitung import Vorverarbeitung
from python_files.Modellauswertung import Modellauswertung
from python_files.Modellauswertung_Dask import Modellauswertung_Dask
```

```python
data_loader = Data_Loader()
vorverarbeitung_train = Vorverarbeitung(data_loader.data_train)
vorverarbeitung_train.set_columns_type(data_loader.data_original)
data_train, weights_train = vorverarbeitung_train.get_data_and_weights()
vorverarbeitung_test = Vorverarbeitung(data_loader.data_test)
vorverarbeitung_test.set_columns_type(data_loader.data_original)
data_test, weights_test = vorverarbeitung_test.get_data_and_weights()
```

```python
X_train = data_train.drop(columns=['income'])
y_train = data_train['income']
X_test = data_test.drop(columns=['income'])
y_test = data_test['income']

modellauswertung = Modellauswertung(X_train, X_test, y_train, y_test)
modellauswertung.evaluate_all_combinations_to_csv("results_orginal_dataset")
```
